```python
import numpy as np
import matplotlib.pyplot as plt
A = []
B = []
C=[]
c = np.pi
x_0 = 1
for k in range (1, 15):
    x_1= x_0+((1+np.cos(x_0))/np.sin(x_0))
    A.append(x_1)
    B.append((x_1 - c)**2) 
    C.append(np.log((x_1 - c)**2))
    x_0=x_1
print(A)
print(B)
plt.plot(C)
 
 
```

```python
import numpy as np
import matplotlib.pyplot as plt
A = []
B = []
C=[]
c = np.sqrt(2)
x_0 = 1
for k in range (1, 10):
    x_1= 0.5*(x_0+2/x_0)
    A.append(x_1)
    B.append((x_1 - c)**2)
    C.append(np.log((x_1 - c)**2))
    x_0=x_1
print(A)
print(B)
plt.plot(C)
```
